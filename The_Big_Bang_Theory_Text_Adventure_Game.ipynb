```python
#DocString Explaining the Premise of the game.
"""
DocString:

A) Introduction: This game is a reiteration of a scene from my favorite TV 
Show, The Big Bang Theory! In this particular scene we can see the protagonist 
Sheldon Cooper come across a retro text based computer game from the 1980s.
I plan to replicate the scenario as best I can using Python.
The game in the show allows for any inputs from the player but I will try to
make the inputs fairly simple for this particular assignment.

Please watch this video to view the scene of the show I am refrencing:
https://www.youtube.com/watch?v=8al0RzkDqgM

B)Flow of Events:
> Start
> Scene 2
> Scene 3
> End

C) Known Bugs/Errors:
None.

"""
    
import random #Importing the necessary packages

#Potential Answer Choices
answer_yes = ["YES", "Y", "yes", "y", "Yes"]
answer_no = ["NO", "N", "no", "n", "No"]



#Defining Start
def start():
    print("""
          The forest of darkness is a mystical land who's existence has always
          been a mystery. You are a sailor who spent several years
          wandering the high seas in search of this forest. You are on your
          adventure when the weather suddenly changes and you are caught in
          a terrible storm! The next few moments are a total blur when your
          boat suddenly crashes onto a dark island with barren black trees. 
          Your ears are filled with howls and cries from animals all over the
          place! This is it...you have found the long lost Forest of Darkness.
          Do you think you have what it takes to end the darkness and bring 
          light into this land?
          """)  
    #Input to begin the game.
    input(prompt = "< Press ENTER to begin. > \n")
    #Creating a list for options
    List_Thirsty = ['thirsty', 'hungry', 'getting impatient']
    
    #Implementing choice options
    thirsty = random.choice(List_Thirsty).lower()
    
    #Printing the opening statemnt of the game.
    print(f"""
    You are entering the forest surrounded by wild animals who look like they
    are {thirsty} for some human blood! You wander the land to try to find the
    cause of its darkness. The sounds of animals around you terrifies you
    and you begin pacing. Suddenly you find yourself in front of giant metal
    door,the sound of the bushes rustling near you intensifies.
    The animals smell you and are nearing your location.
    
    Do you open the door?
    """)
#Providing an input space
    choice = input (prompt = ">>").lower()
    #Implementing Nested Condition

    if choice.lower() in answer_yes:
        print("The door is really heavy! You have to push harder!\n")
        count = 3 #soft coding information
        while count > 0:  #implementing while loops
            print("PUSH")
            count -= 1    
            input(prompt = "Press ENTER to Push Harder!")
            
    #Printing result   
        print("""\nThe door opens! You manage to rush inside and close it
        behind you as the animals approach! The door closes at just the right
        time! Phew! That was a close one!\n""")    
        return Scene_2() #Advancing to scene 2
           
     #Alternative response scenario   
    elif choice.lower() in answer_no:
        print("""You try to find another way around but the animals catch 
        up to you and attack you! You are helpless and unable to survive!""")
        return fail() #Implementing the fail function
    
   # Wrong Input Scenario
    else:
        print("Invalid entry. Please try again.")
        input(prompt = "Press ENTER to continue.\n")
        return start() #Restarting due to Error in input
     

#Defining Scene 2
def Scene_2():
    #Opening Scenario for Scene 2
    print("""
    \n You are Safe...For Now! You managed to enter the door only to see a damp
    land full of small trees and bushes. The floor looks soft and muddy but it
    hasn't been raining. You wonder why the floor is damp as you begin 
    walking forward taking small and cautious steps. You barely make it past
    a few feet when you suddenly realize what you're walking on! Your eyes
    widen in disbelief and before you can react your foot starts sinking.
    QUICKSAND! You are now in a battle to free yourself before the land
    engulfs you whole!
    
    What do you do?
    
    1) Try to wiggle yourself out by using your hands as support.
    2) Grab a nearby branch and pull yourself up.
    3) Wait for your inevitable demise.
    """)
    
    choice = input(prompt = ">>").lower() #input space for user

#Nested Condition
    if '1' in choice or 'wiggle' in choice or 'hand'in choice:
        print("""You realize what a terrible idea this is as your hands are
        now also stuck in the Quicksand! You are left there helpless and
        unable to do anything. You have no choice but to surrender and wait
        for death.""")
        return fail()
    
    #Alternative Scenario based on resonse 2
    elif '2' in choice or 'branch' in choice or 'pull' in choice:
        print("""\nIt's working! You managed to get a good grip and pull
        yourself up with all your strength! You get up on the tree and your
        feet are finally free! You breathe a sigh of relief!\n""")
        return Scene_3() #Advancing to Scene 3
    
    #Alternative Scenario based on resopnse 3
    elif '3' in choice or 'wait' in choice:
        print("That's it? You give up? You are now dead.")
        return fail()
    
    #Wrong input scenario
    else:
        print("Wrong input. Please try again.")
        return Scene_2()

#Defining Scene 3
def Scene_3():
    
    #Opening scenario for Scene 3
    print("""\nYou begin swinging from branch to branch trying to avoid the
    quicksand. You can see the floor starting to dry up and a cave far in
    the distance. You go towards the cave and see that the land is completely
    dry. You carefully step down and walk towards the cave. You hear some
    sounds and you try to throw a rock inside the cave to see if there's
    something there. Oh No! A huge troll walks out and he looks angry!
    He starts sprinting towards you to attack you.
    
    What do you do?

          1) Run
          2) Attack it!""") #choices for the user
    
    choice = input(prompt= ">>").lower() #Input space for the user
    
     #Scenario for response 1
    if '1' in choice or 'run' in choice:
        print("You really think you can outrun a hungry troll? No you can't!")
        return fail() 
    
    #Scenario for Response 2
    elif '2' in choice or 'Attack' in choice:
        print("You go to grab a weapon from your weapons archive:")
        
        #Creating a list for weapon choices
        weapons = ['Sword', 'Axe', 'Machette', 'Spiked Maul']
        
        #Creating a random shuffle for weapon choices
        weapons_answer = random.shuffle(weapons)
        #Creating a random choice for weapons
        weapons_answer = random.choice(weapons).lower()
        
        #Input for weapons answer
        input_weapons = input(prompt = f"""
        Do you use a Sword, Axe, Machette, Spiked Maul, Rock or Rope?\n
        >>""").lower()
        
        #Nested Condition to support the random function
        if input_weapons == weapons_answer: 
            print(f"""You grab a {weapons_answer}!""")
            troll_health = 5 #soft coding information
            while troll_health > 0: #Another while loop
                input(prompt = "Press ENTER to ATTACK!")
                #Decreasing health by 1 everytime user hits Enter
                troll_health -= 1 
            
   #Printing Troll Health to show user how much longer they need to attack.

                print(f"Troll Health: {troll_health}") #Dynamic Printing
    
            #Creating list for Random choices
            List_amazing = ['amazing', 'awesome', 'incredible']
            
            #Implementing Random choice to choose an option
            Amazing = random.choice(List_amazing).lower()
           
            print(f"""
        Great job! That last strike proved fatal as the troll falls
        to its knees before biting the dust. The troll perishes before your 
        very eyes and you see a ray of light penetrating the clouds! The 
        darkness descends and light finally takes over the forest of darkness 
        freeing it from the troll's reign of terror!
        
        You are {Amazing}!!\n""") #Dynamic printing 
        
        
        #End of Game
        
        elif input_weapons != weapons_answer:
            print(f"""That weapon is not effective! You died!
            The correct weapon was {weapons_answer}""")
            return fail() #fail function for wrong input
        else:
            print("Wrong Input. Try Again!")
            return Scene_3() #Returning to Scene 3
    else:
        print("Wrong Input. Try Again!")
        return Scene_3() #Returning to Scene 3
    
#Defining Fail
def fail():
    print("YOU HAVE FAILED!\n")
    input(prompt = 'Press ENTER to exit.\n') #GAME OVER
    

start()
    








```
